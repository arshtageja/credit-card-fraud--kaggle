```python
from sklearn.model_selection import train_test_split
```

```python
data1=data.sample(frac=0.1)
```

```python
data1.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
target=data1["Class"]
```

```python
data1=data1.drop(["Class"],axis=1)
```

```python
data1.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data1, target, test_size=0.3)
```

```python
df=clf.fit(X_train,y_train)
```

```python
y_result=df.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_result)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_result)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier()
```

```python
df1=clf1.fit(X_train,y_train)
```

```python
result1=df1.predict(X_test)
```

```python
accuracy_score(y_test,result1)
```
